### 윤년을 판단하는 함수 만들기

##### 문제
연도가 주어졌을 때, 윤년이면 1, 아니면 0을 리턴하는 함수를 작성하시오.

윤년은 연도가 4의 배수이면서, 100의 배수가 아닐 때 또는 400의 배수일 때이다.

예를 들어, 2012년은 4의 배수이면서 100의 배수가 아니라서 윤년이다. 1900년은 100의 배수이고 400의 배수는 아니기 때문에 윤년이 아니다. 하지만, 2000년은 400의 배수이기 때문에 윤년이다.

##### 입력
첫째 줄에 연도가 주어진다. 연도는 1보다 크거나 같고, 4000보다 작거나 같은 자연수이다.

##### 리턴
첫째 줄에 윤년이면 1, 아니면 0을 리턴한다.


예제 입력 1
2000

예제 리턴 1
1

예제 리턴 2
2011

예제 리턴 2
0

##### 조건: 함수명은 다음과 같이 쓴다. 
``leap_year``

```python
def leap_year(v):
  list1 = []
  x = 0
  while x <= 4000:
    x += 4
    list1.append(x)
  list2 = []
  y = 0
  while y <= 4000:
    y += 400
    list2.append(y)
  list3 = []
  z = 0
  while z <= 4000:
    z += 100
    list3.append(z)
  
  if v in list1:
    if v in list2:
      print(1)
      return 1
    elif v in list3:
      print(0)
      return 0
    else:
      print(0)
      return 1
  else:
    print(0)
    return 0
```
