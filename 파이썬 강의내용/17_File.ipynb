## File
다른 사람들이 정리해놓은 정보를 활용해서 프로그래밍 문제를 해결해야하는 경우가 존재하기에 파이썬 코드 상에서 외부 파일을 어떻게 여는지, 수정하는지, 만드는지를 배우도록 하겠습니다.<br>
프로그래밍 언어에서는 파일을 파일을 연 경우에는 닫아주는 코드를 작성해야합니다.<br>
- open(파일명, 모드): 모드에 따라 파일을 열고 어느 변수에 저장합니다.
    - r: 읽기 / 파일이 없을 때 오류 발생 / 파일이 있을 때 읽기 / 디폴트 값
    - w: 쓰기 / 파일이 없을 때 새로 생성 / 파일이 있을 때 새로 생성
    - a: 추가 / 파일이 없을 때 새로 생성 / 파일이 있을 때 추가
    - x: 배타적 추가 / 파일이 없을 때 새로 생성 / 파일이 있을 때 오류 발생
- close(): 열었던 파일을 닫습니다.
- with: with 구문의 들여쓰기가 시작할 때 파일을 열고 들여쓰기가 끝나면 파일을 자동으로 닫아줍니다.

### w, a모드와 관련된 함수
- write(): 파일에 어떤 내용을 쓸 때 사용합니다.


### r모드와 관련된 함수
- read(): 파일 전체를 읽습니다.
- readline(): 파일을 한줄씩 읽습니다. -> 어느 위치를 읽고 있는지 기억합니다.
- readlines(): 파일의 각 줄이 요소인 list를 리턴합니다.


```python
# 쓰기
# write()
f = open("새파일.txt", "w")          # 쓰기 모드

for i in range(1, 11):
    data = f"line {i}\n"
    f.write(data)                   # 파일에 쓰기

f.close()
```

```python
# 읽기
# 한줄 읽기 readline()
f = open("새파일.txt", "r")         # 읽기 모드
line = f.readline()                # 1번째 줄 
print(line)
line = f.readline()                # 2번째 줄
print(line)

# f 는 파일의 어느 위치까지 읽기 / 쓰기 했는지를 기억하고 있다.

f.close()

# 모든 라인을 읽어오려면?

# 방법 1 : readline() 만으로도 가능
print()

f = open("새파일.txt", "r")
while True:
    line = f.readline()           # 더 이상 읽을 라인이 없는 경우 none 리턴
    if not line: break            # 다 읽었으면 while 종료  
    print(line, end = "")
f.close()

# 방법 2 : readlines()
print()

f = open("새파일.txt", "r")
lines = f.readlines()             # 각각의 line들을 담은 list를 리턴
for line in lines:
    print(line, end="")

f.close()

# 방법 3 : for 사용
print()
f = open("새파일.txt", "r")
for line in f:                    # 파일 객체 f 에서 한 라인씩 읽어들임
    print(line, end="")
f.close()

# 방법 4 : read() 사용
print()

f= open("새파일.txt", "r")
data = f.read()                   # 파일 전체를 읽어들임
print(data)
f.close()
```

```python
# 추가
f = open("새파일.txt", "a")       # 추가 모드
for i in range(11, 20):
    data = "%d Line appended\n" %i
    f.write(data)                # 기존 파일 내용의 끝에서부터 쓰기 시작

f.close()

print("append한 후 결과")
f = open("새파일.txt", "r")
data = f.read()
print(data)
f.close()
```

```python
# with 구문
# 자동으로 close 해준다.
with open("새파일.txt", "r") as f:    # 블럭이 끝나면 f를 자동으로 close 해준다.
    data = f.read()
    print(data)
```

```python
""" 실습 문제
파일의 단어 빈도수 구하기

문제
주어진 텍스트 파일의 알파벳을 소문자로 바꾼 뒤 글자수가 2개 이상이고 빈도수가 2회 이상인 단어만 카운트 하시오


출력 예시
the 1642
and 872
to 729
it 595
she 553
of 514
said 462
you 411
alice 398
in 369
...

* hint 딕셔너리를 활용하세요.
"""
```

```python
# alice 데이터 사용하기
with open('alice30.txt', 'r') as f:
    data = f.read()
    data = data.lower()
    for ch in data:
        if not ('a' <= ch < chr(ord('z')+1)):
            data = data.replace(ch, ' ')
    result = {}
    for word in data.split():
        if word not in result.keys(): # not result.get(word)
            result[word] = 0
        result[word] += 1
    
    # 내림차순
    item = sorted(result.items(), key= lambda x: x[1], reverse = True)
    for key, value in item:
        print(key, value)
```
