## Comprehension
집합 데이터를 더 간편하게 작성하기 위해 ``간략하게`` 집합 데이터를 만드는 코드 구문

### List comprehension
[표현식 for 변수 in 반복가능객체(집합데이터) (if 조건)]
- if 조건문이 있다면 반복문에서 변수에 값을 할당한 다음 if 조건문에서 값을 판별합니다.

````python
# [1, 2, 3 ,4] -> [3, 6, 9, 12]
# 1. list ecomprehension을 사용하지 않는 경우
temp = [1, 2, 3, 4]
result = []
for x in temp:
    result.append(x*3)
print(result)

# 2. list comprehension을 사용하는 경우
result = [x*3 for x in temp]
print(result)
````

````python
# [1, 2, 3, 4] -> [5, 15]
# 1. list comprehension을 사용하지 않는 경우
temp = [1, 2, 3, 4]
result = []
for x in temp:
    if x % 2:
        result.append(x * 5)
print(result)

# 2. list comprehension을 사용하는 경우
result = [x*5 for x in temp if x%2]
print(result)
````

````python
# 실습 문제 ["1월", "2월", ..., "12월"] list comprehension 사용해서 리스트 만들기
result = [str(x) + '월' for x in range(1, 13)]
print(result)

result = [f'{x}월' for x in range(1, 13)]
print(result)
````

### 이중 list의 list comprehension
```python
[표현식 for 변수1 in 반복 가능 객체1 (if 조건1)
     for 변수2 in 반복 가능 객체2 (if 조건2)
     for 변수3 in 반복 가능 객체3 (if 조건3)
     for 변수4 in 반복 가능 객체4 (if 조건4)
 ...
]
```

````python
# 구구단 문제 풀이
result = [dan for dan in range(2, 10)]
print(result)

# 방법 1
# for dan in range(2, 10):
#     for mul in range(1, 10):
#         print(dan, mul, dan*mul)

result = [f'{dan} x {mul} = {dan*mul}' for dan in range(2, 10)
              for mul in range(1, 10)
         ]
for x in result:
    print(x)
````

````python
# 방법 2
result = [[f'{dan} x {mul} = {dan * mul}' for mul in range(1, 10)] for dan in range(2, 10)]
for x in result:
    print(f'{x[0][0]}단')
    for y in x:
        print(y)
````

### Set comprehension
{표현식 for 변수 in 반복가능객체 (if 조건)}

````python
# Set comprehension 기본 구조
result = [num % 3 for num in range(10)]
print(result) # [0, 1, 2, 0, 1, 2, 0, 1, 2, 0]

result = {num % 3 for num in range(10)}
print(result)
````

### Dictionary Comprehension
{key:value for 변수 in 반복가능객체 (if 조건)}

````python
# cf) 딕셔너리 형변환 함수: dict()
# Dictionary comprehension 기본 구조
result = { i:i**2 for i in range(5)}
print(result) # {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
````

### 실습 문제
어떤 문장이 주어졌을 때, 해당 문장이 알파벳 소문자인 경우에 <br>
그 알파벳을 key로 하고, value는 그 알파벳이 문자에 등장하는 횟수를 저장하는 딕셔너리를 <br>
dictionary comprehension을 통해 만들어라.

````python
# 난이도 중
# count()
data = 'aaaabb1acAC'
# result = {'a':5, 'b':2, 'c':1}
result = {alpha: data.count(alpha) for alpha in data if 'a' <= alpha and alpha <= 'z'}
print(result)

# 난이도 상
# count(), lower()
# result = {'a':6, 'b':2, 'c':2}
result = {alpha: data.lower().count(alpha) for alpha in data if 'a' <= alpha and alpha <= 'z'}
print(result)
````
