## Conditional Statement
조건문은 특정한 조건에 해당될(참일) 때만 동작하도록 합니다.<br>
if ~ else ~의 구조로 작성하며 여러번 중첩(nested)해서 사용합니다.<br>
파이썬에서의 조건문은 들여쓰기를 해야합니다. (Tab or space * 4) <br>
이를 지키지 않는 경우는 Syntax 오류가 발생합니다.

### True/False
일반적으로 조건은 bool 타입을 사용해 참/거짓을 판별합니다.<br>
- True: 참
- False: 거짓

```python
# if문의 기본 형태
if True:
    print('조건문이 참1')
if False:
    print('조건문이 참2')
```

```python
# if else문의 기본 형태
if True:
    print('조건문이 참1')
else:
    print('조건문이 거짓1')
    
if False:
    print('조건문이 참2')
else:
    print('조건문이 거짓2')
```

```python
# if ~ elif ~ else ~ 문의 기본 형태
'''
if 첫번째 조건:
    A
elif 두번째 조건:
    B
elif 세번째 조건:
    C
else:
    Final
'''
condition1, condition2, condition3 = False, False, False
if condition1:
    print('첫번째 조건이 참')
elif condition2:
    print('두번째 조건이 참')
elif condition3:
    print('세번째 조건이 참')
else:
    print('모든 조건이 거짓')
```

### 다른 데이터 타입의 bool 활용
다른 데이터 타입의 경우에도 조건문에서 True/False의 역할을 할 수 있습니다.
1. str
    - '': False
    - 그 외의 문자열: True
2. int
    - 0: False
    - 그 외의 정수: True
3. float
    - 0.0: False
    - 그 외의 실수: True

<b>=> 프로그래밍 문제에서 다음과 같은 방식으로 자주 조건문을 판별하니 이를 잘 기억해두시기를 바랍니다.</b>

```python
# if 조건문에서의 str
print(bool('')) # False
print(bool(' ')) # True
print(bool('0')) # True
print(bool('True')) # True

if '':
    print('첫번째 조건문이 참')
elif '0':
    print('두번째 조건문이 참')
else:
    print('모든 조건문이 거짓')
    
# if 조건문에서의 int
print(bool(0)) # False
print(bool(-1)) # True

# if 조건문에서의 float
print(bool(0.0)) # False
print(bool(0.0000000001)) # True

# ?
print(bool(int('0'))) # False
```

### 논리/관계 연산자
논리/관계 연산자를 통해 참과 거짓을 더욱 다양하게 표현할 수 있습니다.
논리연산자
- and: True and True일 때만 True
- or: False or False일 때만 False
- not: True일 때 False, False일 때 True

관계연산자
- \>: 왼쪽이 오른쪽 값보다 더 클 때 True
- <: 왼쪽이 오른쪽 값보다 더 작을 때 True
- \>=: 왼쪽이 오른쪽 값보다 더 크더나 같을 때 True
- <=: 왼쪽이 오른쪽 값보다 더 작거나 같을 때 True
- ==: 왼쪽과 오른쪽 값이 같을 때 True
- !=: 왼쪽과 오른쪽 값이 같지 않을 때 True

```python
# 논리 연산자
# and
print(True and True) # True
print(False and True)
print(True and False)
print(False and False)

print(True and True and True and True and False) # True and False -> False

print('--'*50)
# or
print(True or True)
print(True or False)
print(False or True)
print(False or False) # False

print(False or False or False or False or True) # True
print(0 and 0 or True) # True
```

```python
# shortcut
# print(2/0) ZeroDivisionError
print(True or 2/0) # True
print(False or 3) # 3
print(True and 3) # 3
print(0 and True) # 0 or False
print(0 or 0 and True) # 연산자 우선순위 고려 / and > or 
```

```python
# 관계 연산자
print(2 > 3) # False
print(3 >= 3) # True
print(2 == 2) # True
print(2 != 2) # False

# 정수와 실수의 ==: 데이터 타입이 다르더라도 값을 기준으로 판단
print(2 == 2.0) # True
print(2 is 2.0) # False
print(2 is 2) # True
```

```python
# 문자열간의 크기 비교
# - 사전순으로 크기 비교
print('a' > 'b') # False
print('a' >= 'b') # False
print('a' != 'b') # True

# ord: 문자의 아스키 코드 값을 반환합니다.
print(ord('A'))
print(ord('B'))
print(ord('a'))

# chr: 아스키 코드 값을 문자로 반환합니다.
print(chr(65))
print(chr(98))
```

```python
# 알파벳 검증
alpha = input()
if ('a' <= alpha and alpha <= 'z') or ('A' <= alpha and alpha <= 'Z'):
    print(f'{alpha}는 알파벳입니다.')
else:
    print(f'{alpha}는 알파벳이 아닙니다.')
```

```python
'''실습 문제 2
두 정수 A와 B가 주어졌을 때, A와 B를 비교하는 프로그램을 작성하시오.

입력
첫째 줄에 A와 B가 주어진다. A와 B는 공백 한 칸으로 구분되어져있다.

출력
- A가 B보다 큰 경우에는 '>'를 출력한다.
- A가 B보다 작은 경우에는 '<'를 출력한다.
- A가 B와 같은 경우에는 '=='를 출력한다.
'''
A, B = map(int, input().split())
if A > B:
    print('>')
elif A < B:
    print('<')
else:
    print('==')
```

```python
'''
실습 문제 2

학생들에게 점수를 입력받아
점수가 100점 이하 90점 이상인 경우에는 A를 출력하고,
점수가 90점 미만 80점 이상인 경우에는 B를 출력하고,
점수가 80점 미만 70점 이상인 경우에는 C를 출력하고,
점수가 70점 미만 60점 이상인 경우에는 D를 출력하고,
그 외의 점수에는 F를 출력하는 프로그램을 만드시오.

조건: 점수는 0이상 100 이하의 정수 값을 가집니다.

입력 예시 1
90

출력 예시 1
A

입력 예시 2
75

출력 예시 2
C

입력 예시 3
13

출력 예시 3
F
'''

# if 구문만 사용해서 문제 풀이
score = int(input())
if score >= 90:
    print('A')
if score < 90 and score >= 80: # 80 <= score < 90 -> 파이썬식 스타일
    print('B')
if score < 80 and score >= 70:
    print('C')
if score < 70 and score >= 60:
    print('D')
if score < 60:
    print('F')
    
# if ~ elif ~ else 구문을 활용해서 문제 풀이
if score >= 90:
    print('A')
elif 80 <= score:
    print('B')
elif 70 <= score:
    print('C')
elif 60 <= score:
    print('D')
else:
    print('F')
```
