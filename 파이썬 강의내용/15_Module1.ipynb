## Module
모듈은 특정한 목적에 대해 유용한 변수나 함수, 클래스 등을 모아놓은 파일을 의미합니다.<br>
모듈은 파이썬 파일(.py)로 존재합니다.

### 모듈 불러오기
모듈을 사용하기 위해 ``import``를 사용해야합니다.<br>
import를 사용하는 데에는 크게 두 가지 방법이 있습니다.<br>
- 1. import 모듈
- 2. from 모듈 import 함수나 변수

##### import 모듈
- 해당 방법은 모듈에 저장된 모든 변수나 함수, 클래스 등을 사용할 때 사용합니다.
- 하나의 모듈에서 여러 개의 변수나 함수, 클래스를 사용할 때 편리하게 사용할 수 있습니다.
- 하나의 함수만 사용할 경우에는 낭비가 있을 수 있습니다.

##### from 모듈 import 함수나 변수
- 해당 방법은 모듈에 저장된 특정한 함수(들)를 사용할 때 사용합니다.
- 하나의 모듈에서 적은 변수나 함수, 클래스를 사용할 때 편리하게 사용할 수 있습니다.
- 여러 개의 함수나 변수를 사용할 때는 많이 입력해야해서 가독성이 떨어지고 어떤 함수를 불러왔는지 헷갈릴 수 있고, 변수 명이 겹칠 우려가 있습니다.
    - cf) 여러 개의 함수를 불러올 때: from 모듈 import 함수1, 함수2

```python
# import 모듈
import math
print(math.pi)


# from 모듈 import 함수나 변수
from math import pi
print(pi)
```

#### 1. math 모듈
수학에 관련된 값과 함수를 제공하는 모듈입니다.
- math.pi: 원주율 pi
- math.ceil(): 천장 함수(올림)로 정수로 올림처리합니다.
- math.floor(): 바닥 함수(내림)로 정수로 내림처리합니다.
- math.trunc(): 정수값만 남깁니다.
- math.sqrt(): 제곱근 함수로 주어진 수의 제곱근을 구합니다.
- math.pow(): 제곱함수로 주어진 수의 제곱을 구합니다.

```python
# math 모듈
import math

# 원주율
print(math.pi)

# math.ceil()
print(math.ceil(3.14)) # 4

# math.floor()
print(math.floor(3.14)) # 3

# math.trunc()
print(math.trunc(3.14)) # 

# trunc와 floor 차이는?
print(math.floor(-3.14), math.trunc(-3.14)) # -4 -3 -> 양수일 때는 차이가 없으나 음수인 경우 차이가 존재한다.

# math.sqrt()
print(math.sqrt(25)) # 5.0

# math.pow()
print(math.pow(2, 3)) # 8.0

# math.pow()와 **의 차이는?
print(2 ** 3) # 8 -> 정수
```

#### 2. random 모듈
난수를 생성하는 모듈입니다.
- random.randint(start, end): start 이상 end 이하의 정수 중 하나를 생성합니다.
- random.randrange(end): 0 이상 end 미만의 정수 중 하나를 생성합니다.
- random.randrange(start, end): start 이상 end 미만의 정수 중 하나를 생성합니다.
- random.randrange(start, end, step): start부터 step만큼 end 미만의 정수 중 하나를 생성합니다.
- random.random(): 0 이상 1 미만의 범위에서 임의의 실수를 생성합니다.
- random.choice(sequencial datatype): 순서가 있는 자료형에 속한 요소 중에서 하나를 임의로 반환합니다.
- random.sample(sequencial datatype, count): 순서가 있는 자료형에 속한 요소 중에서 count만큼의 요소를 추출합니다.
- random.shuffle(sequencial datatype): mutable하고 순서가 있는 자료형의 순서를 랜덤하게 재배치합니다.
- random.seed(value): 시드를 주어 random값이 변하지 않도록 합니다.

```python
# random 모듈
import random

# random.seed()
# random.seed(123)



# random.randint()
print(random.randint(1, 10))

# random.randrange()
print(random.randrange(10))
print(random.randrange(1, 10))
print(random.randrange(1, 10, 2))

# randint와 randrange가 헷갈릴 때는? 자료구조를 활용해 테스트 해보자.
randintList = [0 for _ in range(11)]
randrangeList = [0 for _ in range(11)]
for _ in range(10000):
    idx1 = random.randint(0, 10)
    idx2 = random.randrange(0, 10)
    randintList[idx1] += 1
    randrangeList[idx2] += 1

print(randintList)
print(randrangeList) # 10 미만의 값만 가짐

# random.random()
print(random.random())

# 50% 확률 계산
if random.random() < 0.5:
    print('성공')
    
# random.choice()
seasons = ['sprint', 'summer', 'fall', 'winter']
print(random.choice(seasons))

# random.sample()
print(random.sample(seasons, 2))

# random.shuffle()
random.shuffle(seasons)
print(seasons)

# shuffle 사용 시 주의할 점
seasons2 = 'sprint', 'summer', 'fall', 'winter'
# random.shuffle(seasons2) TypeError -> mutable한 sequencial datatype만 사용 가능하다.
```

```python
''' 실습 문제 1
로또 추첨 프로그램

문제
1. 1에서 45 사이의 모든 정수를 순서대로 pot 리스트에 저장합니다.
2. 다음 과정을 6번 반복합니다.
    - pot 리스트를 무작위로 섞어줍니다.
    - pot 리스트의 마지막 숫자를 하나만 빼서 jackpot 리스트에 저장합니다.
    - 2초 동안 잠시 멈춥니다.
3. jackpot 리스트의 모든 요소를 오름차순 정렬합니다.
4. jackpot 리스트의 모든 요소를 출력합니다.

출력 예시
1번째 당첨번호는 38입니다.
2번째 당첨번호는 42입니다.
3번째 당첨번호는 25입니다.
4번째 당첨번호는 23입니다.
5번째 당첨번호는 40입니다.
6번째 당첨번호는 7입니다.
이번 당첨번호는 [7, 23, 25, 38, 40, 42] 입니다.

* 자세한 값은 달라질 수 있습니다.

문제 해결하신 분들을 위한 심화 문제
- 로또를 뽑기 전에 사용자로부터 6개의 숫자를 입력받아 실제 로또 번호와 비교해 상금이 얼마인지 출력하는 프로그램을 만들어라!
1등: 6개 맞기 -> 3천만원
2등: 5개 맞기 -> 2백만원
3등: 4개 맞기 -> 5만원

'''

from random import shuffle
pot = [x for x in range(1, 46)]
jackpot = []
for i in range(1, 7):
    shuffle(pot)
    x = pot.pop()
    jackpot.append(x)
    print(f'{i}번째 당첨번호는 {x}입니다.')
jackpot.sort()
print(f'이번 당첨번호는 {jackpot} 입니다.')
```

#### 3. time 모듈
time 모듈은 시간 처리에 관한 모듈입니다.
- time.time(): 1970년 1월 1일 0시 0분 0초부터 현재까지 경과된 시간을 반환합니다.
- time.ctime(time): 인수로 전달된 시간을 형식을 갖춰 반환합니다.
- time.strftime(지시자): 형식을 설정해 날짜를 원하는 대로 문자열로 만들어 반환할 수 있습니다.
- time.sleep(second): second만큼 시스템을 중지시킵니다.

<br>
* datetime 모듈은 교재 참고

```python
# time 모듈
import time

# time.time()
print(time.time())

# time.ctime()
print(time.ctime(1231231))
print(time.ctime(time.time()))

# time.strftime()
print(time.strftime('%Y-%m-%d %a %H:%M:%S')) # 교재 참고

# time.sleep()
print(time.sleep(2))
print('2초?')

# 프로그램 시간 체크할 때 time함수 자주 사용
startTime = time.time()
time.sleep(2)
endTime = time.time()
elapsedTime = endTime - startTime
print(elapsedTime)
```
