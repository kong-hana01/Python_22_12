### 파이썬 집합 데이터
- list: 순서가 있다. 중복을 허용한다. mutable
- tuple: 순서가 있다. 중복을 허용한다. immutable
- set: 순서가 없다. 중복을 허용하지 않는다.
- dict: key와 value 쌍으로 구성된다. 순서가 없다.

### dict
- dictionary는 key-value 쌍으로 저장되는 데이터 집합입니다.
- dictionary는 순서가 없기 때문에 인덱싱이 불가능하지만 key를 통한 검색이 가능합니다.

### dict의 기본 구조
- dictionary는 {}를 통해 데이터를 표현합니다.
```python
{key1:value1, key2:value2, key3:value3 ...}
```

````python
# 딕셔너리의 기본구조
student = {}
print(student, type(student))

student = {
    'name':'최한빈', # key1:value1 사이에 : 적어주기, key-value 쌍 간에 ,로 구분해주기
    'e-mail':'choi@naver.com'
}
print(student)
````

### dictionary의 데이터 검색
1. key 값을 활용합니다.
2. get()함수를 활용합니다.

get 함수를 활용할 경우에는 예외적인 상황에서도 동작 가능하도록 처리가 가능합니다.

````python
# 딕셔너리의 데이터 검색
# 방법 1: key값 사용
print(student['name']) # 최한빈
# print(student['id']) # KeyError     

# 방법 2: get(key) 사용
print(student.get('name')) # 최한빈
print(student.get('id')) # None
print(student.get('id', 10)) # 10
````

### dictionary의 데이터 수정/추가/삭제
1. 데이터의 수정은 key 값을 적어주고, 할당 연산자(=)를 통해 수정할 값을 입력합니다.
2. 데이터의 추가는 만들 key값을 적어주고, 할당 연산자를 통해 값을 추가합니다.
3. 데이터의 삭제는 del함수를 사용합니다.

````python
student = {
    'name':'최한빈', # key1:value1 사이에 : 적어주기, key-value 쌍 간에 ,로 구분해주기
    'e-mail':'choi@naver.com'
}
print(student)

# 데이터의 수정
student['name'] = '한빈최'
print(student)

# 데이터의 추가
student['address'] = '김포'
print(student)

# 데이터의 삭제
del(student['e-mail'])
print(student)
````

### dictionary의 key/value
딕셔너리의 key값은 해쉬가 가능해야합니다. (집합데이터는 key값으로 사용이 불가능합니다.) <br>
다만, 딕셔너리의 value는 어떤 데이터든 상관이 없습니다.<br>

###### keys(), values(), items()
- keys(): 딕셔너리의 모든 key를 반환합니다.
- values(): 딕셔너리의 모든 value를 반환합니다.
- items(): 딕셔너리의 모든 key와 value 쌍을 반환합니다.

````python
dict1 = {
    1: "Haha",
    'two': {
        3.14 : 'pi',
        'pi' : 3.14,
        'score' : [100, 200, 300]
    },
    False : [10, 20, 30, 40],
    # [1, 2] : 'Hello' TypeError: unhashable type: 'list'
}
print(dict1[1]) # Haha
print(dict1['two'], type(dict1['two'])) # 딕셔너리 출력 {3.14: 'pi', 'pi': 3.14, 'score': [100, 200, 300]} <class 'dict'>
print(dict1['two']['score'], type(dict1['two']['score'])) # [100, 200, 300] <class 'list'>
print(dict1['two']['score'][0]) # 100
````

````python
# keys()
print(dict1.keys(), type(dict1.keys()))
print(dict1['two'].keys())

# values()
print(dict1.values())

# items()
print(dict1.items())

# items() 활용
for key, value in dict1.items():
    # 코드 동작
    print(key, value)
````

### 딕셔너리의 in 연산자
``key``의 존재 여부를 확인할 때 사용합니다.

````python
print(student)
print('name' in student) # True
print('한빈최' in student) # False
print('id' in student) # False
````
