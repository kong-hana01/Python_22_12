## while 문
while문은 주어진 조건이 참인 경우에 반복하는 구문입니다.<br>
기본적인 구조는 다음과 같습니다.<br>

```python
while (조건):
    조건이 참일 때 동작할 구문 1
    조건이 참일 때 동작할 구문 2
    조건이 참일 때 동작할 구문 3
    ...
```
조건이 참일 때 동작할 구문 내에서 원하는 목적을 달성했으면 while 조건문에서 나올 수 있는 과정이 필요합니다.<br>
ex) while i < 10인 경우 i의 값을 반복적으로 더해주는 과정이 필요합니다.

````python
while True: # 무한 루프
    print(1)
````

````python
i = 1
while i < 10:
    print(i)
    i += 1
````

````python
# for문과의 비교
for i in range(1, 10):
    print(i)
print('-' * 100)

# while문은 동일한 출력을 다양한 방법으로 할 수 있다.
i = 1
while i < 10:
    print(i)
    i += 1
print('i의 값:', i)

print('-' * 100)
i = 0
while i < 9:
    i += 1
    print(i)
print('i의 값:', i)
````

### while문 작성시 주의 사항
while 문에서는 조건이 참인 경우, 동작을 계속해서 반복하기 때문에 무한루프에 빠질 수 있습니다.<br>
무한루프에 빠진 경우 어떤 점이 잘못됐는지 확인해봅시다.<br>
가장 기본적인 것은 조건문을 빠져나오기 위한 조건을 출력해보는 것입니다.

````python
i = 0
while i < 10:
    print(i)
    i += 1
````

### break, continue, pass
조건문에서는 반복문을 나오거나, 현재 반복중인 부분만 넘기거나, 반복적인 구문 중 어느 일부분만 넘기도록 별도로 지정해놓은 예약어가 있습니다.<br>
- break: 반복문을 나올 때 사용합니다.
- continue: 현재 반복 중인 것을 넘기고 다음 반복으로 넘어갑니다.
- pass: if 구문으로 감싼 부분만 넘기고 현재 반복중인 부분 중 나머지 부분을 동작합니다.

````python
# break
for i in range(10):
    if i == 2:
        print('break')
        break
    print(i)

print()

# continue
for i in range(10):
    if i == 2:
        print('continue')
        continue
    print(i)
    
print()

# pass
for i in range(10):
    if i == 2:
        print('pass')
        pass
    print(i)
````
